```python
#Cargo datos
df_train = pd.read_csv('blackFriday_train.csv')
```

```python
#Columnas que posee 
df_train.columns
```

##Diccionario de los campos
**User_ID**: Numero unico que identifica a cada usuario

**Gender**: Indica el género del usuario ('M' para masculino y 'F' para femenino).

**Age**: Indica la edad del usuario en grupos de edad predefinidos ('0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+').

**Occupation:** Número que indica la ocupación del usuario, codificado en una lista predefinida.

**City_Category**: Indica la categoría de la ciudad en la que reside el usuario ('A' para ciudades de nivel alto, 'B' para ciudades de nivel medio y 'C' para ciudades de nivel bajo).

**Stay_In_Current_City_Years**: Indica el número de años que el usuario ha estado viviendo en la ciudad actual ('0' para menos de 1 año, '1' para 1 año, '2' para 2 años, '3' para 3 años y '4+' para más de 4 años).

**Marital_Status**: Indica el estado civil del usuario ('0' para soltero y '1' para casado).

**Product_Category_1**: Indica la categoría de producto principal a la que pertenece el producto, codificado en una lista predefinida.

**Product_Category_2**: Indica la categoría de producto secundaria a la que pertenece el producto, codificado en una lista predefinida.

**Product_Category_3**: Indica la categoría de producto terciaria a la que pertenece el producto, codificado en una lista predefinida.

**Purchase**: Indica el monto total gastado por el usuario en la compra del producto.
