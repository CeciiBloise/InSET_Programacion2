2.19. Obtener el país con mayor cantidad de usuarios cuya edad sea
mayor a 50.


```python
#Agrupo los usuarios por país, filtro los que tienen una edad mayor a 50 y cuento el número de usuarios en cada grupo
df[df['age'] > 50].groupby('country')['name'].count()
```

```python
#Ordeno los grupos de mayor a menor y selecciono el primero
df[df['age'] > 50].groupby('country')['name'].count().sort_values(ascending=False).index[0]
```

2.20. Obtener el país con mayor promedio de edad.


```python
#Agrupo los usuarios por país y calculo el promedio
df.groupby('country')['age'].mean()
```

```python
#Ordeno los grupos de mayor a menor promedio y selecciono el primero
df.groupby('country')['age'].mean().sort_values(ascending=False).index[0]
```

2.21. Mostrar en consola el país con más hombres.

```python
#Ordeno los grupos de mayor a menor y selecciono el primero
df[df['gender'] == 'male'].groupby('country')['gender'].count().sort_values(ascending=False).index[0]
```
