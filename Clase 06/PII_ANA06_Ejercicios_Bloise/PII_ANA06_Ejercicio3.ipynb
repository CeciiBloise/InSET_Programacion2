```python
#Importo las librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Cargo el archivo
df_usuarios=pd.read_csv("usuarios.csv")
df_diabetes=pd.read_csv("diabetes.csv")
```

#Graficos para diabetes.csv

**Grefico de la distribucion de la Glucosa**

```python
sns.histplot(data=df_diabetes, x='Glucosa', kde=True)
```

**Grafico de la correlación entre variables numéricas**

```python
sns.heatmap(df_diabetes.corr(), annot=True)
```

**Grafico de barras de "Clase"**

```python
sns.countplot(data=df_diabetes, x='Clase')
```

**Grafico de la dispersión de la Edad y la Insulina**

```python
sns.scatterplot(data=df_diabetes, x=' Edad', y=' InsulinaSéricaDe2HRS', hue='Clase')
```

**Grafico de pastelcon la distribución de usuarios por país**

```python
#Cuento la cantidad de usuarios por país
usuarios_por_pais = df_usuarios['country'].value_counts()

#Selecciono los primeros 5 países con más usuarios
top_5_paises = usuarios_por_pais.head(5)

#Agrupo los usuarios de otros países bajo una categoría "Otros"
otros_paises = pd.Series({'Otros': usuarios_por_pais[5:].sum()})

#Concato los 5 países principales y "Otros"
usuarios_por_pais = pd.concat([top_5_paises, otros_paises])

#Creo el grafico
plt.pie(usuarios_por_pais, labels=usuarios_por_pais.index, autopct='%1.1f%%')
plt.title('Distribución de usuarios por país')
plt.show()
```

**Grafico de barras de la cantidad de usuarios por género y país**

```python
counts = df_usuarios.groupby(['country', 'gender']).size().unstack()
counts.plot(kind='bar', stacked=True)
plt.xlabel('País')
plt.ylabel('Cantidad de usuarios')
plt.title('Cantidad de usuarios por género y país')
plt.show()
```

**Grafico de pastel de la proporción de géneros en el conjunto de datos**

```python
counts = df_usuarios['gender'].value_counts()
plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')
plt.title('Proporción de géneros en el conjunto de datos')
plt.show()
```

**Grafico de burbujas de usuarios**

```python
# Seleccionar las columnas a utilizar
x = df_usuarios['age']
y = df_usuarios.index
sizes = df_usuarios['age']
colors = df_usuarios['gender'].apply(lambda x: 'blue' if x == 'male' else 'red')

# Crear el gráfico de burbujas
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, alpha=0.5)

# Agregar etiquetas y título
ax.set_xlabel('Edad')
ax.set_ylabel('Índice de usuario')
ax.set_title('Gráfico de burbujas de la distribución de usuarios por edad y ubicación')

plt.show()


```
